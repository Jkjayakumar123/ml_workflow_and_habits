```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
#Task1-Categorical encoding
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
# 2. City → One-Hot Encoding (drop one column)
df=pd.get_dummies(df,columns=['City'],drop_first=True)
# 3. Size → Custom Label Encoding (ordered
size_map={'Small':0,'Medium':1,'Large':2}
df['Size'] = df['Size'].map(size_map)


# Task 2 — Feature Scaling

# Using RobustScaler because Fare has outliers (e.g., 450)
# RobustScaler uses median and IQR, so it is not affected by extreme values

scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

# Output
print(df)

```
